```python
#pseudocode from lesson

def is_balanced(s):
    st = stack
    for char in s:
        if char in ["[","(","{"]:
            st.push(char)
        else:
            if st.empty(): 
                return False
            top=st.pop()
            if (top="[" and char!="]") or  (top="(" and char!=")") or (top="{" and char!="}"):
                return False
    return st.empty()
            
```

```python
#stack class from http://aliev.me/runestone/BasicDS/ImplementingaStackinPython.html
class Stack:
     def __init__(self):
         self.items = []

     def isEmpty(self):
         return self.items == []

     def push(self, item):
         self.items.append(item)

     def pop(self):
         return self.items.pop()

     def peek(self):
         return self.items[len(self.items)-1]

     def size(self):
         return len(self.items)
```

```python
#https://stepik.org/lesson/41234/step/1?discussion=1218281&unit=19818
    


from queue import LifoQueue


def check_bracket(string: str) -> bool:
    stack = LifoQueue()

    for s in string:
        if s in "([{":
            stack.put(s)
        elif s in "}])":
            b = stack.get()
            if b == "[" and s != "]":
                return False
            elif b == "{" and s != "}":
                return False
            elif b == "(" and s != ")":
                return False
        else:
            raise ValueError(f"String '{s}' is not a bracket")

    return True


if __name__ == "__main__":
    assert check_bracket("([](){([])})")
    assert not check_bracket("()[]}")
    assert not check_bracket("{{[()]]")


import pytest

@pytest.mark.parametrize("string,result", [("([](){([])})", True), ("()[]}", False), ("{{[()]]", False)])
def test_check_bracket(string, result):
    assert check_bracket(string) == result
   

```
