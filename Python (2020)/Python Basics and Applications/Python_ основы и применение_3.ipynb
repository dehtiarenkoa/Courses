```python
#works as needed
import requests
import json

client_id = 'bcfc2561934e8d5b4b07'
client_secret = '98f6ede39448c74a440e4a39059e6d11'

# инициируем запрос на получение токена
r = requests.post("https://api.artsy.net/api/tokens/xapp_token",
                  data={
                      "client_id": client_id,
                      "client_secret": client_secret
                  })

# разбираем ответ сервера
j = json.loads(r.text)

# достаем токен
token = j["token"]

# создаем заголовок, содержащий наш токен
headers = {"X-Xapp-Token" : token}
# инициируем запрос с заголовком

with open("dataset_24476_4.txt") as f:
    sortable_name = "_"
    d = {}
    while sortable_name!="":
        sortable_name = f.readline().strip()
        if sortable_name=="":
            continue
        #print(sortable_name)
        r = requests.get("https://api.artsy.net/api/artists/"+sortable_name, headers=headers)
        # разбираем ответ сервера
        j = json.loads(r.text)
        print(j["name"])
        print(j["birthday"])
        d[j["sortable_name"]]=j["birthday"]
for i in d:
    print(i)
print()
#d = {k:v for k, v in sorted(d.items(), key=lambda item: item[1], reverse=False)}
d = {k:v for k, v in sorted(d.items(), key=lambda item: (item[1], item[0]))}
print("_d_")
with open("dataset_24476_4_a.txt", "w", encoding="utf8") as f:
    for i in d:    
        f.write(i+"\n")
```
